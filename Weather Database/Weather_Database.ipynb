```python
lats = np.random.uniform(-90,90,2000)
long = np.random.uniform(-180,180,2000)
```

```python
lat_long = zip(lats, long)
```

```python
cities = []
for coor in lat_long:
    city = citipy.nearest_city(coor[0],coor[1]).city_name
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
rcd = 0
sets = 1
for city in cities:

    rcd+=1
    print(f'record {rcd} of set {sets} | {city}')
    if rcd == 50:
        rcd = 0
        sets += 1
```
